```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# all columns are displayed
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
df = pd.read_csv("bmw_cleaned.csv")
```

```python
df.head()
```

```python
# Displays how many cars of each color there are.
pd.pivot_table(df, index = ['model_key','paint_color'], values = 'price', aggfunc = 'count').sort_values('model_key', ascending = False)
```

```python
# Displays the price of different color for vehicles
pd.pivot_table(df, index = ['model_key','paint_color'], values = 'price').sort_values('model_key', ascending = False)
```

```python
pd.pivot_table(df, index = ['paint_color'], values = 'price').sort_values('price', ascending = False)
```

```python
pd.pivot_table(df, index = ['fuel'], values = 'price').sort_values('price', ascending = False)
```

```python
pd.pivot_table(df, index = ['model_key', 'car_type'], values = 'price').sort_values('price', ascending = False)
```

```python
# Shows price difference between BMW MPowered cars and non MPowered
pd.pivot_table(df, index = ['MPower_yn'], values = 'price').sort_values('price', ascending = False)
```
